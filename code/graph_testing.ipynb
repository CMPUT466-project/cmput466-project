```python
import matplotlib.pyplot as plt
seq_len = [5, 10, 15, 25, 50, 100, 200]
train_acc = [0.4929, 0.6272, 0.6652, 0.7010, 0.7516, 0.8208, 0.8556]
vali_acc = [0.2651, 0.2576, 0.2613, 0.2621, 0.2479, 0.2566, 0.2632]
plt.xlabel('sequence length')
plt.ylabel('training accuracy')
plt.plot(seq_len, train_acc)
plt.show()
plt.xlabel('sequence length')
plt.ylabel('validation accuracy')
plt.plot(seq_len, vali_acc)
plt.show()
```
